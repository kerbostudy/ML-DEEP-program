#### 강화학습 : Reinforcement Leanring
####                   에이전트 (Agent : 사람)
####    상태(state)        보상(reward)         행동(action)  
####                     환경 (Environment)
#### 마르코프      프로세스 : Markov          Process, M P
#### 마르코프 보상 프로세스 : Markov Reward   Process, MRP
#### 마르코프 결정 프로세스 : Markov Decesion Process, MDP  
####                      => 벨만 기대 방정식: Bellman expectation equation

#### 몬테카를로 트리 탐색 : 게임 시뮬레이션을 이용하여 가장 가능성이 높아 보이는 방향으로 행동을 결정하는 탐색 방법
#### 선택 -> 확장 -> 시뮬레이션 -> 역전파  : 계속적인 반복함

```python
#12.5.2 몬테카를로 트리 검색을 적용한 틱택토 게임 구현하기
# 보드 생성 : 게임 좌표 3*3 위치에 1,2,3,4,5,6,7,8,9 포진
boarder = { '1': ' ' , '2': ' ' , '3': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '7': ' ' , '8': ' ' , '9': ' ' }

board_keys = []

for key in boarder:
    board_keys.append(key)
```

```python
# 화면 출력 함수 정의
def visual_Board(board_num):            #  게임을 진행할 때마다 Update 된 보드를 출력해야 함, visual_Board() 함수를 호출하여 쉽게 보드를 출력함
    print(board_num['1'] + '|' + board_num['2'] + '|' + board_num['3'])
    print('-+-+-')
    print(board_num['4'] + '|' + board_num['5'] + '|' + board_num['6'])
    print('-+-+-')
    print(board_num['7'] + '|' + board_num['8'] + '|' + board_num['9'])
```

```python
# Agent(플레이어, 사람)가 보드를 이동하여 게임을 진행하는 메인 함수 정의
def game():
    turn = 'X'                                                                    # X 마크
    count = 0

    for i in range(8):
        visual_Board(boarder)
        print("당신 차례입니다," + turn + ". 어디로 이동할까요?")                 # 게임 순번으로 어디에 위치할지 질문?
        move = input()
        if boarder[move] == ' ':
            boarder[move] = turn
            count += 1
        else:
            print("이미 채워져있습니다.\n어디로 이동할까요?")                     # 이미 게임지 진행된 위치에 또 위치하는 것은 불가능!
            continue

        if count >= 5:
            if boarder['1'] == boarder['2'] == boarder['3'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['4'] == boarder['5'] == boarder['6'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['7'] == boarder['8'] == boarder['9'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['1'] == boarder['4'] == boarder['7'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['2'] == boarder['5'] == boarder['8'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['3'] == boarder['6'] == boarder['9'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['1'] == boarder['5'] == boarder['9'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

            elif boarder['3'] == boarder['5'] == boarder['7'] != ' ':
                visual_Board(boarder)
                print("\n게임 종료.\n")
                print(" ---------- " +turn + "가 승리했습니다. -----------")
                break

        if count == 9:
            print("\n게임 종료.\n 게임을 다시")
            print("동점입니다")


        if turn =='X':
            turn = '0'
        else:
            turn = 'X'

if __name__ == "__main__":
    game()
```
