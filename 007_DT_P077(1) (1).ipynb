#### # 데이터를 1차로 분류한 후
#### 정보 획득(information gain) : 각 영역의 순도(homegeneity)는 증가 & 불순도(impurity -max 05) 와 불확실성(uncertainty)은 감소하는 방형으로 학습을 진행

#### 결정 트리에서 불확실성을 계산하는 2가지 방법
#### 1) 엔트로피(entropy: 0~0.5) : 확률변수의 불확실성을 수치로 나타낸것
#### 2) 지니계수(Gini Index) :



```python
# 데이터 전처리
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]     # 승객의 생존여부 예측을 위한 Feature
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = df.dropna()                                                             # dropna(), 값이 없는 데이터 삭제

X = df.drop('Survived', axis=1)
y = df['Survived']                                                          # 'Survived'를 예측 레이블로 사용
```

```python
# Training / Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Model (DT) ,  fit():Model training
from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)             # 모델 훈련
```

```python
# 모델 예측
y_predict = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```

# cross-matrics
#  TP     FN
#  FP     TN

```python
# Cross-Matrics , 성능평가
from sklearn.metrics import confusion_matrix
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=['Predicted Not Survival', 'Predicted Survival'],
    index=['True Not Survival', 'True Survival']
)
```
