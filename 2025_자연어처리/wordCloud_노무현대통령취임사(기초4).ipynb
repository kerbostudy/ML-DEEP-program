```python
from wordcloud import WordCloud
from collections import Counter
from konlpy.tag import Okt
import matplotlib.pyplot as plt
```

```python
with open('/content/nmhp.txt', 'r') as f:
  speech_text = f.read()
```

```python
print(speech_text)
```

```python
# 2. Okt 형태소 분석기 초기화
okt = Okt()
```

```python
# 3. 명사 추출
nouns = okt.nouns(speech_text)
```

```python
print(nouns)
```

```python
nouns = [word for word, pos in okt.pos(speech_text) if pos == 'Noun']
print(nouns)
```

```python
# 4. 불용어 정의 (필요에 따라 불용어 목록을 수정할 수 있습니다.)
stopwords = []
```

```python
# 5. 불용어 제거 및 빈도수 계산
filtered_nouns = [noun for noun in nouns if noun not in stopwords]
noun_counts = Counter(filtered_nouns)
```

```python
print(len(noun_counts))
```

```python
# 6. 가장 빈번한 단어 100개 추출
top_100 = dict(noun_counts.most_common(100))
```

```python
print(top_100)
```

```python
from gensim.models import Word2Vec

# 2. 명사 리스트를 Word2Vec 모델 학습에 사용할 데이터 형태로 변환
# 전체 텍스트에서 추출된 명사들을 하나의 문장으로 취급합니다.
word2vec_input = [nouns]

# 3. Word2Vec 모델 학습
model = Word2Vec(word2vec_input, vector_size=100, window=5, min_count=5, sg=0)

# 3. 모델 저장 (선택 사항)
model.save("word2vec.model")

# 4. 단어 벡터 확인
print(model.wv['국민'])  # '오늘' 단어의 벡터 출력
```

```python
# '국민'과 유사한 단어 찾기
similar_words = model.wv.most_similar('국민', topn=10)  # 상위 10개 단어

# 결과 출력
for word, similarity in similar_words:
    print(f"{word}: {similarity}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

from gensim.models import Word2Vec

# Word2Vec 모델 로드
model = Word2Vec.load("word2vec.model")

# 모든 단어를 노드로 추가
graph = nx.Graph()
for word in model.wv.key_to_index:
    graph.add_node(word)

# 단어 간 유사도가 특정 임계값 이상인 경우 연결선 추가
threshold = 0.6  # 유사도 임계값 설정
for word1 in model.wv.key_to_index:
    for word2 in model.wv.key_to_index:
        if word1 != word2:
            similarity = model.wv.similarity(word1, word2)
            if similarity > threshold:
                graph.add_edge(word1, word2, weight=similarity)

# 네트워크 시각화
plt.figure(figsize=(15, 15))  # 그래프 크기 조정
pos = nx.spring_layout(graph, k=0.3)  # 노드 위치 설정, k값 조정으로 노드 간 거리 조절
nx.draw_networkx_nodes(graph, pos, node_size=1000, node_color="skyblue")  # 노드 그리기
nx.draw_networkx_edges(graph, pos, width=[d['weight'] * 2 for u, v, d in graph.edges(data=True)], alpha=0.5)  # 연결선 그리기, 가중치에 따라 두께 조절, 투명도 조절
nx.draw_networkx_labels(graph, pos, font_family="NanumBarunGothic", font_size=10)  # 레이블 그리기, 한글 폰트 설정
plt.axis("off")  # 축 숨기기
plt.title("Word2Vec 단어 네트워크", fontproperties=fontprop)  # 제목, 한글 폰트 설정
plt.show()
```
