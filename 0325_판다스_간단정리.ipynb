```python
## 딕셔너리 타임 데이어 key=열이름, value는 시퀀스형 데이터 타임 넣고 각 열의 데이터로 만듦
# from pandas import Series, DataFrame

raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'age'       : [42, 52, 36, 24, 73],
            'city'      : ['San Francisco', 'Baltimore', 'Miami', 'Douglas', 'Boston']}

df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'city'])
df
```

```python
df.describe()
```

```python
df = pd.DataFrame(raw_data, columns = ['age', 'city'])
df
```

```python
# 속성이 넘치는 경우, 즉 데ㅣ터가 존쟂하기 않는 열을 추가하면 해당 열에는 NaN값이 생김
df2 = DataFrame(raw_data, columns = ["first_name","last_name","age", "city", "debt"] )
df2
```
